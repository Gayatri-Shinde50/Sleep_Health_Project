```python
person_by_gender=hc.groupby('patient_gender')['patient_id'].count().reset_index()
person_by_gender
```

```python
plt.figure(figsize=(6,6))
plt.pie(person_by_gender['patient_id'],labels=person_by_gender['patient_gender'],autopct='%1.1f%%',startangle=90)
plt.title('No of Patients per Gender')
plt.show()
```

```python
grouped_race=hc.groupby('patient_race')['patient_sat_score'].mean().reset_index()
grouped_race
```

```python
plt.figure(figsize=(14,6))
plt.bar(grouped_race['patient_race'],grouped_race['patient_sat_score'])
plt.xlabel('Patient race')
plt.ylabel('Average Patient sat score')
plt.title('Average Patient sat score by Patient race')
plt.show()
```

```python
def categorized_age(patient_age):
    if patient_age>=1 and patient_age<20:
        return 'Child'
    elif patient_age>=20 and patient_age<30:
        return 'Young'
    elif patient_age>=30 and patient_age<40:
        return 'Adult'
    elif patient_age>=40 and patient_age<50:
        return 'Middle Age'
    elif patient_age>=50:
        return 'Old age'
    else:
        return 'Unknown'
```

```python
hc['Age Group']=hc['patient_age'].apply(categorized_age)
```

```python
person_per_age_group=hc.groupby('Age Group')['patient_id'].count().sort_values(ascending=False).reset_index()
person_per_age_group
```

```python
plt.figure(figsize=(6,6))
plt.pie(person_per_age_group['patient_id'],labels=person_per_age_group['Age Group'],autopct='%1.1f%%',startangle=90)
plt.title('No of Patients per Age Griup')
plt.show()
```

```python
avg_waittime=hc.groupby('department_referral')['patient_waittime'].mean().reset_index()
avg_waittime
```

```python
plt.figure(figsize=(14,6))
plt.bar(avg_waittime['department_referral'],avg_waittime['patient_waittime'])
plt.xlabel('Department Refferal')
plt.ylabel('Average Patient waittime')
plt.title('Average Patient waittime by department refferal')
plt.show()
```
